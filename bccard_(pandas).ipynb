```python
cvs_df = bccard_df[bccard_df['TP_BUZ_NO'] == 4010] # 편의점(4010) 데이터만 추출
cvs_df
```

```python
cvs_df['AMT'].sum() # 전체 편의점 소비
```

```python
gangnam_df = bccard_df[bccard_df['CTY_RGN_NM'] == '강남구'] # 강남구 데이터만 추출
gangnam_df
```

```python
gangnam_df['AMT'].sum() # 강남구 전체 소비액
```

```python
gangnam_cvs = bccard_df[(bccard_df['CTY_RGN_NM'] == '강남구') & (bccard_df['TP_BUZ_NO'] == 4010)] # 강남구에 있는 편의점 데이터만 추출(두 조건 동시 만족)
gangnam_cvs
```

```python
gangnam_cvs['AMT'].sum() # 강남구 편의점 총 소비액
```

```python
resident_amt = gangnam_cvs[gangnam_cvs['CSTMR_CTY_RGN_NM'] == '강남구']['AMT'].sum() # 강남구 편의점 소비액 중 강남구 거주자와 비거주자의 소비액 비교
resident_amt # 강남구 편의점 (깅남구 거주자) 소비액
```

```python
non_resident_amt = gangnam_cvs[gangnam_cvs['CSTMR_CTY_RGN_NM'] != '강남구']['AMT'].sum() # 강남구 편의점 소비액 중 강남구 거주자와 비거주자의 소비액 비교
non_resident_amt # 강남구 편의점 (비거주자) 소비액
```
