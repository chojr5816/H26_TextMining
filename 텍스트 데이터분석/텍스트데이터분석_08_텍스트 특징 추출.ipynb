# 텍스트 특징 추출 (TF-IDF)

### 1. 샘플 코퍼스로 연습

```python
sample_corpus = [
    '자연어처리 강의를 시작하겠습니다.',
    '자연어처리는 재미있습니다.',
    '밥을 먹고 강의를 듣고 있습니다.',
    '이번 자연어처리 강의는 한국어 자연어처리입니다.'
]
```

```python
import sklearn
```

```python
from konlpy.tag import Okt
def my_tokenizer(text):
    return Okt().nouns(text)
```

```python
# TfidfVectorizer 객체 생성
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=my_tokenizer)
# from sklearn.feature_extraction.text import CountVectorizer
# vectorizer = CountVectorizer(tokenizer=my_tokenizer)
# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(sample_corpus)
vectorizer.get_feature_names_out()
#특징 벡터 추출
sample_dtm = vectorizer.transform(sample_corpus)
type(sample_dtm)
print(vectorizer.get_feature_names_out())
print(sample_dtm.toarray())
```

### 2. 다음 영화 리뷰로 연습

```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
# from sklearn.feature_extraction.text import TfidfVectorizer

okt = Okt()

def my_tokenizer(text):
    return okt.nouns(str(text))

# 1. 데이터 불러오기
df = pd.read_csv('./data/daum_movie_review.csv')
# 행의 개수 제한 (1000개)
df_short = df.head(1000)
my_stopwords = ['영화', '이', '것', '하는', '정말', '진짜', '보고', '점', '좀', '그', '수', '할', '있는', '때', '보는']

vectorizer = CountVectorizer(tokenizer=my_tokenizer,
                                stop_words=my_stopwords,
                                max_features=30, # 열의 개수 제한 (30개)
                                min_df=2,
                                max_df=0.8) 

                              
# # vectorizer = TfidfVectorizer(tokenizer=my_tokenizer,
#                              stop_words=my_stopwords,
#                              max_features=1000,
#                              min_df=2,
#                              max_df=0.8,
#                              token_pattern=None)
vectorizer.fit(df_short['review'])
vectorizer.get_feature_names_out()
#특징 벡터 추출
review_dtm = vectorizer.transform(df_short['review'])
type(review_dtm)
print(vectorizer.get_feature_names_out())
print(review_dtm.toarray())
```
