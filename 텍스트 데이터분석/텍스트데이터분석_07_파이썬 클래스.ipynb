# 파이썬 클래스

```python
class Account:
    # self.balance = 0 # 멤버 변수 balance 선언이 안된다.(클래스 변수 balance 선언)
    balance = 100 # 클래스 변수 balance 선언
    # def __init__(self): # 멤버 변수 balance 초기화 -> 변수 선언
    #   self.balance = 0
    def deposit(self, amount): # 입금 메서드
        self.balance += amount
    def withdraw(self, amount): # 출금 메서드
        self.balance -= amount

a1 = Account() # 객체 생성
a1.deposit(100)
a1.withdraw(50)
print('a1 : ', a1.balance)
print('Account : ', Account.balance)   

print(Account.balance.__hash__) 
print(a1.balance.__hash__)
```

```python
class Account:
    def __init__(self): # 멤버 변수 balance 초기화 -> 변수 선언
        self.__balance = 0
    def deposit(self, amount): # 입금 메서드
        self.__balance += amount
    def withdraw(self, amount): # 출금 메서드
        self.__balance -= amount
    def get_balance(self):
        return self.__balance

a1 = Account() # 객체 생성
a1.deposit(100)
a1.withdraw(50)
print('a1 : ', a1.get_balance())
```
